```python
import numpy as np
import pandas as pd
from sklearn.metrics import recall_score, confusion_matrix, classification_report, precision_score, accuracy_score, f1_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.datasets import load_diabetes, load_iris, make_classification, make_regression
from sklearn.model_selection import train_test_split

#Zadanie1
print('Zadanie 1')
iris = load_iris()
x_train_iris, x_val_iris, y_train_iris, y_val_iris = train_test_split(
    iris.data, iris.target, train_size=0.3, random_state=42)

knn_iris = KNeighborsClassifier(n_neighbors=2)
knn_iris.fit(x_train_iris,y_train_iris)

#1.A
y_pred_iris = knn_iris.predict(x_val_iris)

#1.B
print('Prognozowane:', y_pred_iris[:15])
print('Walidacyjne:', y_val_iris[:15])

#1.C
accuracy_iris = accuracy_score(y_val_iris,y_pred_iris)

#1.D
print('Dokladnosc (walidacyjny):', accuracy_iris)

#1.E
train_accuarcy_iris = accuracy_score(y_train_iris, knn_iris.predict(x_train_iris))
print('Dokladnosc (treningowy):', train_accuarcy_iris)
print('')

#Zadanie2
print('Zadanie 2')
#2.A
precision_iris = precision_score(y_val_iris, y_pred_iris, average='weighted')
print('Precyzja:', precision_iris)

#2.B
recall_iris = recall_score(y_val_iris, y_pred_iris, average='weighted')
print('Czulosc:', recall_iris)

#2.C
f1score_iris = f1_score(y_val_iris, y_pred_iris, average='weighted')
print('F1 score:', f1score_iris)

#2.D
conf_mat_iris = confusion_matrix(y_val_iris, y_pred_iris)
print('Macierz pomylek:', conf_mat_iris)

#2.E
class_report_iris = classification_report(y_val_iris, y_pred_iris)
print('Raport:', class_report_iris)

#Zadanie3
print('Zadanie 3')
diabetes = load_diabetes()
x_train_diabetes, x_val_diabetes, y_train_diabetes, y_val_diabetes = train_test_split(
    diabetes.data, diabetes.target, train_size=250, random_state=42)

knn_diab = KNeighborsRegressor(n_neighbors = 2)
knn_diab.fit(x_train_diabetes,y_train_diabetes)

#3.A
y_pred_diab = knn_diab.predict(x_val_diabetes)

#3.B
print('Prognozowane:', y_pred_diab[:10])
print('Walidacyjne:', y_val_diabetes[:10])

#3.C
mae = mean_absolute_error(y_val_diabetes, y_pred_diab)

#3.D
print('MAE (walidacyjny):', mae)
print('Target:', diabetes.target[:5])

#3.E
mea_train = mean_absolute_error(y_train_diabetes, knn_diab.predict(x_train_diabetes))
print('MAE (treningowy):', mae)
print('')

#Zadanie4
print('Zadanie 4')
#4.A
mse_diab = mean_squared_error(y_val_diabetes, y_pred_diab)
print('MSE:', mse_diab)

#4.B
rmse_diab = np.sqrt(mse_diab)
print('RMSE:', rmse_diab)

#4.C
mape_diab = np.mean(np.abs(y_val_diabetes - y_pred_diab) / y_val_diabetes) * 100
print('MAPE:', mape_diab)

#4.D
r2_diab = r2_score(y_val_diabetes, y_pred_diab)
print('R2:', r2_diab)
print('')

#Zadanie5
print('Zadanie 5')
print('Takie samo jak zadanie 1')
print('')

#Zadanie6
print('Zadanie 6')
x_binary, y_binary = make_classification(n_samples=500, n_features=5)
x_train_bin, x_val_bin, y_train_bin, y_val_bin = train_test_split(
    x_binary, y_binary, train_size=0.8, random_state=42)

knn_binary = KNeighborsClassifier(n_neighbors = 2)
knn_binary.fit(x_train_bin, y_train_bin)

#6.A
y_pred_binary = knn_binary.predict(x_val_bin)

#6.B
print('Prognozowane:', y_pred_binary[:15])
print('Walidacyjne:', y_val_bin[:15])

#6.C
accuracy_bin = accuracy_score(y_val_bin, y_pred_binary)

#6.D
print('Dokladnosc (walidacja):', accuracy_bin)

#6.E
accuracy_train_bin = accuracy_score(y_train_bin, knn_binary.predict(x_train_bin))
print('Dokladnosc (treningowa):',accuracy_train_bin) 

print('')

#6.A
precision_bin = precision_score(y_val_bin, y_pred_binary, average='weighted')
print('Precyzja:', precision_bin)

#6.B
recall_bin = recall_score(y_val_bin, y_pred_binary, average='weighted')
print('Czulosc:', recall_bin)

#6.C
f1score_bin = f1_score(y_val_bin, y_pred_binary, average='weighted')
print('F1 Score:', f1score_bin)

#6.D
conf_mat_bin = confusion_matrix(y_val_bin, y_pred_binary)
print('Macierz pomylek:', conf_mat_bin)

#6.E
class_report_bin = classification_report(y_val_bin, y_pred_binary, target_names=["Klasa 0", "Klasa 1"])
print('Klasyfikacja:', class_report_bin)
print('')

#Zadanie7
print('Zadanie 7')
print('')

#Zadanie8
print('Zadanie 8')
x_reg, y_reg = make_regression(n_samples=400, n_features=4)
x_train_reg, x_val_reg, y_train_reg, y_val_reg = train_test_split(
    x_reg, y_reg, train_size=0.8, random_state=42)

knn_reg = KNeighborsRegressor(n_neighbors = 2)
knn_reg.fit(x_val_reg, y_val_reg)

#8.A
y_pred_reg = knn_reg.predict(x_val_reg)

#8.B
print('Prognozowane:', y_pred_reg[:10])
print('Walidacyjne:', y_val_reg[:10])

#8.C
mse_reg = mean_absolute_error(y_val_reg, y_pred_reg)

#8.D
print('MAE (walidacyjny):', mae_reg)
#print('target:', diabetes.target[:5])

#8.E
mae_reg_train = mean_absolute_error(y_train_reg, knn_reg.predict(x_train_reg))
print('MAE (treningowy):', mae_reg_train)
print('')

#8.A
mse_reg = mean_squared_error(y_val_reg, y_pred_reg)
print('MSE:', mse_reg)

#8.B
rmse_reg = np.sqrt(mse_reg)
print('RMSE:', rmse_reg)

#8.C
mape_reg = np.mean(np.abs(y_val_reg - y_pred_reg)/y_val_reg) *100
print('MAPE:', mape_reg)

#8.D
r2_reg = r2_score(y_val_reg, y_pred_reg)
print('R2:', r2_reg)

print('')

#Zadanie9
print('Zadanie 9')

```
