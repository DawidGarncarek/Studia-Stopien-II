```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes, load_iris, make_classification, make_regression
from sklearn.model_selection import train_test_split

#Lista3
#Zadanie 1
#1.1
diabetes = load_diabetes()

#1.2
print('Nazwy kolumn:', diabetes.feature_names)

#1.3
print('Liczba pacjent贸w:', diabetes.data.shape[0])

#1.4
plt.hist(diabetes.target, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Warto zmiennej docelowej')
plt.ylabel('Liczba przypadk贸w')
# plt.show

#Zadanie2
#1.1
iris = load_iris()

#1.2
print('Nazwy kolumn:', iris.feature_names)

#1.3
print('Liczba gatunk贸w:', iris.data.shape[0])

#1.4
unique_species = np.unique(iris.target)
species_names = iris.target_names
print('Poszczeg贸lnych przypadk贸w:', species_names)

#Zadanie3
x_binary, y_binary = make_classification(n_samples=500, n_features=5)
print('Rozmiar wygenerowanegp zbioru binarnego:', x_binary.shape)

#Zadanie4
x_multiclass, y_multiclass = make_classification(n_samples=1000, n_features=10, n_informative=6, n_classes=5)
print('Rozmiar wygenerowanego zbioru dla 5 klasowej klasyfikacji:', x_multiclass.shape)

#Zadanie5
x_regression, y_regression = make_regression(n_samples=400, n_features=4)
print('Rozmiar zbioru dla regresji:', x_regression.shape)

#Zadanie6
x_regressiontwo, y_regressiontwo = make_regression(n_samples=600, n_features=6)
print('Rozmiar zbioru dla drugiej regresji:', x_regressiontwo.shape) 

#Zadanie7
x_train_diabetes, x_val_diabetes, y_train_diabetes, y_val_diabetes = train_test_split(
    diabetes.data, diabetes.target, train_size=250)
print('Rozmiar zbioru trenignowego:', x_train_diabetes.shape)
print('Rozmiar zbioru walidacyjnego:', x_val_diabetes.shape)

#Zadanie8
x_train_iris, x_val_iris, y_train_iris, y_val_iris = train_test_split(
    iris.data, iris.target, train_size=0.7, random_state=42)
print('Rozmiar zbioru trenignowego:', x_train_iris.shape)
print('Rozmiar zbioru walidacyjnego:', x_val_iris.shape)

#Zadanie9
x_train_binary, x_val_binary, y_train_binary, y_val_binary = train_test_split(
    x_binary, y_binary, train_size=0.5, stratify=y_binary)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].hist(y_train_binary, bins=2, color='blue', edgecolor='black')
axs[0].set_title('Histogram zadanie 9 (trenignowy)')
axs[0].set_xlabel('Klasa')
axs[0].set_ylabel('Liczba przypadk贸w')

axs[1].hist(y_val_binary, bins=2, color='orange', edgecolor="black")
axs[1].set_title('Histogram zadanie 9 (walidacja)')
axs[1].set_xlabel('Klasa')
axs[1].set_ylabel('Liczba przypadk贸w')
plt.show()

#Zadanie10
y_regressiontwo_rounded = pd.qcut(y_regressiontwo, q=5, labels=False)

x_train_reg, x_val_reg, y_train_reg, y_val_reg = train_test_split(
    x_regressiontwo, y_regressiontwo, train_size=100, stratify=y_regressiontwo_rounded)

fig, asx = plt.subplots(1, 2, figsize=(12, 5))
asx[0].hist(y_train_reg, bins=20, color='blue', edgecolor='black')
asx[0].set_title('Histogram zadanie 10 (treningowy)')
asx[0].set_xlabel('Zmienna')
asx[0].set_ylabel('Liczba przypadk贸w')

asx[1].hist(y_val_reg, bins=20, color='orange', edgecolor='black')
asx[1].set_title('Histogram zadanie 10 (walidacyjny)')
asx[1].set_xlabel('Zmienna')
asx[1].set_ylabel('Liczba przypadk贸w')
plt.show()

#Lista4
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print(' ')
print(' ')
print(' ')

#Zadanie1
iris = load_iris()
x_train, x_val, y_train, y_val = train_test_split(iris.data, iris.target)

knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train, y_train)

#1.A
y_prediction = knn.predict(x_val)

#1.B
print('Prognozowane klasy:', y_prediction[:15])
print('Rzeczywiste klasy:', y_val[:15])

#1.C
accuracy = accuracy_score(y_val, y_prediction)
print('Dokadno:', accuracy)

#1.D
print('Klasy targetu:', iris.target_names)

#1.E
train_accuracy = accuracy_score(y_train, knn.predict(x_train))
print('Dokadno (trenignowy):', train_accuracy)
print('Dokadno (walidacyjny):', accuracy)

#Zadanie2
#2.A
precision = precision_score(y_val, y_prediction, average='weighted')
print('Precyzja:', precision)

#2.B
recall = recall_score(y_val, y_prediction, average='weighted')
print('Czuo:', recall)

#2.C
fonescore = f1_score(y_val, y_prediction, average='weighted')
print('F1 Score:', fonescore)

#2.D
conf_matrix = confusion_matrix(y_val, y_prediction)
print('Macierz pomyek:', conf_matrix)

#2.E
class_raport = classification_report(y_val, y_prediction)
print('Raport:', class_raport)

#Zadanie3
diabetes = load_diabetes()
x_train_diabetes, x_val_diabetes, y_train_diabetes, y_val_diabetes = train_test_split(
    diabetes.data, diabetes.target, train_size=250)

knn_reg = KNeighborsRegressor(n_neighbors=2)
knn_reg.fit(x_train_diabetes, y_train_diabetes)

#3.A
y_pred = knn_reg.predict(x_val_diabetes)

#3.B
print('Prognozy', y_pred[:10])
print('Rzeczywiste wartoci:', y_train_diabetes[:10])

#3.C
mae = mean_absolute_error(y_val_diabetes, y_pred)
print('MEA:', mae)

#3.D
print('Wartoci target:', diabetes.target[:5])

#3.E
train_mae = mean_absolute_error(y_train_diabetes, knn_reg.predict(x_train_diabetes))
print('MAE (trenignowy):', train_mae)
print('MAE (walidacyjny):', mae)

#Zadanie4
#4.A
mse = mean_squared_error(y_val_diabetes, y_pred)
print('MSE:', mse)

#4.B
rmse = np.sqrt(mse)
print('RMSE:', rmse)

#4.C
mape = np.mean(np.abs((y_val_diabetes - y_pred) / y_val_diabetes)) * 100
print('MAPE:', mape)

#4.D
r2 = r2_score(y_val_diabetes, y_pred)
print('R2:', r2)

#Zadanie5
iris = load_iris()
x_train_i, x_val_i, y_train_i, y_val_i = train_test_split(iris.data, iris.target)

knn_iris = KNeighborsRegressor(n_neighbors=2)
knn_iris.fit(x_train_i, y_train_i)

#5.A
y_pred_i = knn_iris.predict(x_val_i)

#5.B
print('Porgnozowane klasy:', y_pred_i[:15])
print('Rzeczywiste klasy:', y_val_i[:15])

#5.C
#accuracy_iris = accuracy_score(y_val_i, y_pred_i)
#print('Dokadno:', accuracy_iris)

#5.D
print('Klasy targetu:', iris.target_names)

#5.E
#train_accuracy_iris = accuracy_score(y_train_i, knn_iris.predict(x_train_iris))
#print('Dokadno (treningowy):', train_accuracy_iris)
#print('Dokadno (walidacyjny):', accuracy_iris)


```
