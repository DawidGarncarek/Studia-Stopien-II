```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes, load_iris, make_classification, make_regression
from sklearn.model_selection import train_test_split
from collections import Counter
print('')
#Zadanie1
print('Zadanie 1')
diabetes = load_diabetes()
print('Kolumny:', diabetes.feature_names)
print('Liczba pacjentów:', diabetes.data.shape[0])
plt.hist(diabetes.target, bins=20, color='blue', edgecolor='black')
plt.title('Historgam')
plt.xlabel('Zmienna')
plt.ylabel('Liczba przypadkow')
plt.show()
print('')

#Zadanie2
print('Zadanie 2')
iris = load_iris()
print('Kolumny:', iris.feature_names)
unique_species = np.unique(iris.target)
species_names = iris.target_names
print('Liczba gatunków:', species_names)
species_counts = Counter(iris.target)
for species, count in species_counts.items():
    print(f"Gatunek {species_names[species]}: {count} przypadków")

print('')

#Zadanie3
x_binary, y_binary = make_classification(n_samples=500, n_features=5)

#Zadanie4
x_multiclass, y_multiclass = make_classification(n_samples=1000, n_informative=6, n_features=10, n_classes=5)

#Zadanie5
x_reg1, y_reg1 = make_regression(n_samples=400, n_features=4)

#Zadanie6
x_reg2, y_reg2 = make_regression(n_samples=600, n_features=6)

#Zadanie7
x_train_diabetes, x_val_diabetes, y_train_diabetes, y_val_diabetes = train_test_split(
    diabetes.data, diabetes.target, train_size=250, random_state=42)

#Zadanie8
x_train_iris, x_val_iris, y_train_iris, y_val_iris = train_test_split(
    iris.data, iris.target, train_size=0.7, random_state=42)

#Zadanie9
print('Zadanie 9')
x_train_binary, x_val_binary, y_train_binary, y_val_binary = train_test_split(
    x_binary, y_binary, train_size=0.5, stratify=y_binary)

fig, axs = plt.subplots(1, 2, figsize=(12,5))
axs[0].hist(y_train_binary, bins=2, color='blue', edgecolor='black')
axs[0].set_title('Histogram treningowy')
axs[0].set_xlabel('Klasa')
axs[0].set_ylabel('Liczba przypadków')

axs[1].hist(y_val_binary, bins=2, color='orange', edgecolor='black')
axs[1].set_title('Histogram walidacyjny')
axs[1].set_xlabel('Zmienna')

plt.show()
print('')
print('')
#Zadanie10
print('Zadanie 10')
y_reg2_rounded = pd.qcut(y_reg2, q=5, labels=False)

x_train_reg2, x_val_reg2, y_train_reg2, y_val_reg2 = train_test_split(
    x_reg2, y_reg2, train_size=100, stratify=y_reg2_rounded)

fig, axs = plt.subplots(1, 2, figsize=(12,5))
axs[0].hist(y_train_reg2, bins=2, color='blue', edgecolor='black')
axs[0].set_title('Histogram treningowy')
axs[0].set_xlabel('Klasa')
axs[0].set_ylabel('Liczba przypadków')

axs[1].hist(y_val_reg2, bins=2, color='orange', edgecolor='black')
axs[1].set_title('Histogram walidacyjny')
axs[1].set_xlabel('Zmienna')
plt.show()
print('')
```
