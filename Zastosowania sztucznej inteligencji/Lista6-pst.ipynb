```python
import cv2
import matplotlib.pyplot as plt

print("1. Wczytaj z dysku obrazek z użyciem OpenCV")
image_cv = cv2.imread('C:\\Users\\d_gar\\sunflower.jpg')

print("• wyświetl obrazek z użyciem OpenCV")
cv2.imshow('Obrazek - OpenCV', image_cv)
cv2.waitKey(0)
cv2.destroyAllWindows()

print("• wyświetl obrazek z użyciem matplotlib")
image_rgb = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title('Obrazek - Matplotlib')
plt.axis('off')
plt.show()

print("• przekonwertuj obrazek z użyciem OpenCV i wyświetl ponownie z użyciem matplotlib")
image_gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray')
plt.title('Obrazek w skali szarości - Matplotlib')
plt.axis('off')
plt.show()

print(" ")
from PIL import Image
import numpy as np
print("2. Wczytaj z dysku obrazek z użyciem PIL")
image_pil = Image.open('C:\\Users\\d_gar\\meadow.jpg')

print("• wyświetl obrazek z użyciem matplotlib")
plt.imshow(image_pil)
plt.title('Obrazek - Matplotlib (PIL)')
plt.axis('off')
plt.show()

print("• wyświetl obrazek z użyciem OpenCV")
image_pil_cv = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGB2BGR)
cv2.imshow('Obrazek - OpenCV (PIL)', image_pil_cv)
cv2.waitKey(0)
cv2.destroyAllWindows()

print("• przekonwertuj obrazek z użyciem OpenCV i wyświetl ponownie z użyciem OpenCV")
image_pil_gray = cv2.cvtColor(image_pil_cv, cv2.COLOR_BGR2GRAY)
cv2.imshow('Obrazek w skali szarości - OpenCV', image_pil_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()

print("")
print("3. Z użyciem opencv uruchom kamerkę")
camera = cv2.VideoCapture(0)

print("• wyświetl obraz z kamery")

print("• po naciśnięciu klawisza X zapisz obraz z kamery do pliku")

print("• zamknij otwarte okienka")

print("• zamknij połączenie z kamerą")

while True:
    ret, frame = camera.read()
    if not ret:
        break
    cv2.imshow('Kamera', frame)
    
    # Sprawdzenie, czy naciśnięto klawisz 'X'
    if cv2.waitKey(1) & 0xFF == ord('x'):
        cv2.imwrite('kamera_zapis.jpg', frame)  # Zapis obrazu do pliku
        break

camera.release()
cv2.destroyAllWindows()

print("• wyświetl zapisany obraz z użyciem matplotlib")

saved_image = cv2.imread('kamera_zapis.jpg')
saved_image_rgb = cv2.cvtColor(saved_image, cv2.COLOR_BGR2RGB)
plt.imshow(saved_image_rgb)
plt.title('Zapisany obraz')
plt.axis('off')
plt.show()

print(" ")
print("4. wczytaj z dysku obrazek z użyciem OpenCV (obrazek powinien przedstawiać zbliżenie na przynajmniej 2 twarze w okularach)")
image_faces = cv2.imread('C:\\Users\\d_gar\\faces_with_glasses.jpg')

print("• z użyciem metody Haar Cascades wykryj i policz na obrazku twarze w okularach")
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye_tree_eyeglasses.xml')

gray_faces = cv2.cvtColor(image_faces, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray_faces, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

print("• wypisz liczbę wykrytych obiektów")
print(f'Liczba wykrytych twarzy: {len(faces)}')

# Dodatkowo
# Rysowanie prostokątów wokół wykrytych twarzy
for (x, y, w, h) in faces:
    cv2.rectangle(image_faces, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Wyświetlanie obrazu z wykrytymi twarzami
image_faces_rgb = cv2.cvtColor(image_faces, cv2.COLOR_BGR2RGB)
plt.imshow(image_faces_rgb)
plt.title('Wykryte twarze')
plt.axis('off')
plt.show()

print(" ")
print("5. z użyciem opencv uruchom kamerkę")
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
camera = cv2.VideoCapture(0)

print("• wyświetl obraz z kamery")

print("• wyświetl obraz z kamery z zaznaczonymi wykrytymi twarzami (z użyciem metody Haar Cascades)")

print("• w trakcie działania programu wypisuj liczbę aktualnie wykrytych obiektów")
 
print("• po naciśnięciu klawisza X zamknij otwarte okienka")

print("• zamknij połączenie z kamerą")

print("• zapisz do pliku obraz z wykrytymi obiektami")
try:
    while True:
        ret, frame = camera.read()
        if not ret:
            print("Błąd odczytu obrazu z kamery.")
            break
        
        # Konwersja obrazu na skalę szarości
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        # Wykrywanie twarzy
        faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
        
        # Rysowanie prostokątów wokół wykrytych twarzy
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
        # Wyświetlanie liczby wykrytych twarzy w konsoli
        print(f'Liczba wykrytych twarzy: {len(faces)}')
        
        # Wyświetlanie obrazu
        cv2.imshow('Kamera z wykrywaniem twarzy', frame)
        
        # Sprawdzenie, czy użytkownik nacisnął klawisz 'x', aby zakończyć
        if cv2.waitKey(1) & 0xFF == ord('x'):
            # Zapisanie obrazu z wykrytymi twarzami
            cv2.imwrite('kamera_z_twarzami.jpg', frame)
            print("Zapisano obraz z wykrytymi twarzami do pliku.")
            break
except KeyboardInterrupt:
    print("Program przerwany przez użytkownika.")
finally:
    # Zamykanie kamery i okien
    camera.release()
    cv2.destroyAllWindows()
    print("Zamknięto kamerę i wszystkie okienka.")
```
