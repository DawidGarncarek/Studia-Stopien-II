```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes, load_iris, make_classification, make_regression
from sklearn.model_selection import train_test_split
from collections import Counter


# 1.1 Otwórz zbiór
print('1.1 Otwrcie zbioru diabetes:');
diabetes = load_diabetes()
print(' ');

# 1.2 Wypisz nazwy kolumn
print('1.2 Wypisanie nazw kolum:');
print("Nazwy kolumn:", diabetes.feature_names)
print(' ');

# 1.3 Policz ile przebadanych pacjentów opisano w zbiorze
print('1.3 Policzenie przebadanych pacjentów w zbiorze:');
print("Liczba pacjentów:", diabetes.data.shape[0])
print(' ');

# 1.4 Narysuj histogram dla badanej zmiennej
print('1.4 Histogram dla badanej zmiennej:');
plt.hist(diabetes.target, bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram dla zmiennej docelowej (cukrzyca)")
plt.xlabel("Wartość zmiennej docelowej")
plt.ylabel("Liczba przypadków")
plt.show()
print(' ');

# 2.1 Otwórz zbiór
print('2.1 Otwarzcie zbioru iris:');
iris = load_iris()
print(' ');

# 2.2 Wypisz nazwy kolumn
print('2.2 Wypisanie kolumn:');
print("Nazwy kolumn:", iris.feature_names)
print(' ');

# 2.3 Policz i wypisz nazwy badanych gatunków kwiatków
print('2.3 Policzenie i wypisanie nazwy badanych kwiatków:');
unique_species = np.unique(iris.target)
species_names = iris.target_names
print("Nazwy gatunków kwiatów:", species_names)
print(' ');

# 2.4 Policz i wypisz ile przypadków jest w zbiorze dla poszczególnych gatunków
print('2.4 Policzenie i wypisanie ile przypadków jest w zbiorze poszczególnych gatunków:');
species_counts = Counter(iris.target)
for species, count in species_counts.items():
    print(f"Gatunek {species_names[species]}: {count} przypadków")
print(' ');

# 3. Generowanie zbioru dla binarnej klasyfikacji
print('3. Generowanie zbioru dla binarnej klasyfikacji:');
X_binary, y_binary = make_classification(n_samples=500, n_features=5, n_classes=2, random_state=42)
print("Rozmiar wygenerowanego zbioru (binary):", X_binary.shape)
print(' ');

# 4. Generowanie zbioru dla 5 klasowej klasyfikacji
print('4. Generowanie zbioru dla 5 klasowej klasyfikacji:');
X_multiclass, y_multiclass = make_classification(n_samples=1000, n_features=10, n_informative=6, n_classes=5, random_state=42)
print("Rozmiar wygenerowanego zbioru (multiclass):", X_multiclass.shape)
print(' ');

# 5. Generowanie zbioru dla regresji (400 przypadków, 4 kolumny)
print('5. Generowanie zbioru dla regresji (400 przypadków, 4 kolumny):');
X_reg1, y_reg1 = make_regression(n_samples=400, n_features=4, random_state=42)
print("Rozmiar wygenerowanego zbioru (regresja 1):", X_reg1.shape)
print(' ');

# 6. Generowanie zbioru dla regresji (600 przypadków, 6 kolumn)
print('6. Generowanie zbioru dla regresji (600 przypadków, 6 kolumn):');
X_reg2, y_reg2 = make_regression(n_samples=600, n_features=6, random_state=42)
print("Rozmiar wygenerowanego zbioru (regresja 2):", X_reg2.shape)
print(' ');

#7. Podział zbioru cukrzycy na treningowy (250 pacjentów) i walidacyjny
print('7. Podział zbioru na treningowy (250 pacjentów) i walidacyjny: ');
X_train_diabetes, X_val_diabetes, y_train_diabetes, y_val_diabetes = train_test_split(
    diabetes.data, diabetes.target, train_size=250, random_state=42
)
print("Rozmiar zbioru treningowego (diabetes):", X_train_diabetes.shape)
print("Rozmiar zbioru walidacyjnego (diabetes):", X_val_diabetes.shape)
print(' ');

#8. Podział zbioru iris na treningowy (70%) i walidacyjny
print('8. Podział zbioru iris na treningowy (70%) i walidacyjny: ');
X_train_iris, X_val_iris, y_train_iris, y_val_iris = train_test_split(
    iris.data, iris.target, train_size=0.7, random_state=42
)
print("Rozmiar zbioru treningowego (iris):", X_train_iris.shape)
print("Rozmiar zbioru walidacyjnego (iris):", X_val_iris.shape)
print(' ');

#9. Podział zbioru binarnej klasyfikacji (stratyfikacja)
print('9. Podział zbioru binarnej klasyfikacji (stratyfikacja): ');
X_train_binary, X_val_binary, y_train_binary, y_val_binary = train_test_split(
    X_binary, y_binary, train_size=0.5, stratify=y_binary, random_state=42)
print(' ');

# Rysowanie histogramów dla klas
print('Rysowanie histogramów dla klas: ');
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].hist(y_train_binary, bins=2, color='blue', edgecolor='black')
axs[0].set_title("Histogram klas - zbiór treningowy")
axs[0].set_xlabel("Klasa")
axs[0].set_ylabel("Liczba przypadków")

axs[1].hist(y_val_binary, bins=2, color='orange', edgecolor='black')
axs[1].set_title("Histogram klas - zbiór walidacyjny")
axs[1].set_xlabel("Klasa")
plt.show()
print(' ');

# 10. Podział zbioru regresji (stratyfikacja)
# Zaokrąglanie zmiennej docelowej do 5 grup, aby umożliwić stratyfikację
print('10. Podział zbioru regresji (stratyfikacja):');
y_reg2_rounded = pd.qcut(y_reg2, q=5, labels=False)

# Podział zbioru na treningowy i walidacyjny z zachowaniem stratyfikacji
X_train_reg, X_val_reg, y_train_reg, y_val_reg = train_test_split(
    X_reg2, y_reg2, test_size=100, stratify=y_reg2_rounded, random_state=42)

# Rysowanie histogramów dla zmiennej docelowej
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].hist(y_train_reg, bins=20, color='blue', edgecolor='black')
axs[0].set_title("Histogram zmiennej - zbiór treningowy")
axs[0].set_xlabel("Wartość")
axs[0].set_ylabel("Liczba przypadków")

axs[1].hist(y_val_reg, bins=20, color='orange', edgecolor='black')
axs[1].set_title("Histogram zmiennej - zbiór walidacyjny")
axs[1].set_xlabel("Wartość")
plt.show()



```
