
# Wprowadzenie do Seaborn

Seaborn to biblioteka Python służąca do tworzenia atrakcyjnych i informacyjnych wizualizacji danych.
Bazuje na Matplotlib i integruje się z Pandas, co czyni ją idealnym narzędziem do analizy danych tabelarycznych.

W tym notebooku znajdują się:
- Podstawowe przykłady użycia Seaborn,
- Tworzenie różnych typów wykresów,
- Zaawansowane przykłady z wizualizacją danych.

```python

import seaborn as sns
import numpy as np

# Dane
data = np.random.randn(1000)

# Histogram z gęstością
sns.displot(data, kde=True, bins=30)
```

```python

import seaborn as sns
import pandas as pd

# Dane
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [5, 7, 6, 8, 7],
    'category': ['A', 'A', 'B', 'B', 'C']
})

# Scatterplot
sns.scatterplot(data=data, x='x', y='y', hue='category', style='category', s=100)
```

```python

import seaborn as sns
import numpy as np

# Dane
data = np.random.rand(10, 10)

# Heatmap
sns.heatmap(data, annot=True, cmap='viridis', cbar=True)
```

```python

import seaborn as sns

# Dane
data = sns.load_dataset("iris")
print(data)
# Pairplot
sns.pairplot(data, hue='species')
```

```python

import seaborn as sns
import pandas as pd

# Dane
data = pd.DataFrame({
    'category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'value': [10, 15, 8, 12, 14, 18]
})

# Boxplot
sns.boxplot(data=data, x='category', y='value')
```

```python

import seaborn as sns
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

```python

import seaborn as sns

# Dane
data = sns.load_dataset("tips")

# FacetGrid
g = sns.FacetGrid(data, col="time", row="sex", margin_titles=True)
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")

# Load the example tips dataset
tips = sns.load_dataset("tips")
print(tips)
# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
sns.despine(offset=10, trim=True)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark")

# Simulate data from a bivariate Gaussian
n = 10000
mean = [0, 0]
cov = [(2, .4), (.4, .2)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, n).T

# Draw a combo histogram and scatterplot with density contours
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=x, y=y, levels=5, color="w", linewidths=1)
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Tworzenie danych
np.random.seed(10)
data = pd.DataFrame({
    'category': np.repeat(['A', 'B', 'C'], 100),
    'value': np.concatenate([
        np.random.normal(5, 1, 100),  # Dane dla kategorii A
        np.random.normal(7, 1.5, 100),  # Dane dla kategorii B
        np.random.normal(6, 2, 100)   # Dane dla kategorii C
    ])
})
print(data)

# Wykres wiolinowy
sns.violinplot(data=data, x='category', y='value', palette='muted')
#sns.violinplot(data=data, x='category', y='value', inner='point', palette='coolwarm')

```
