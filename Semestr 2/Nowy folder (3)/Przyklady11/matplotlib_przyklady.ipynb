
# Wprowadzenie do Matplotlib

Matplotlib to biblioteka Python służąca do tworzenia wizualizacji danych. 
Oferuje szeroką gamę wykresów, takich jak wykresy liniowe, słupkowe, punktowe, histogramy, a także wykresy 3D.

W tym notebooku znajdują się:
- Podstawowe przykłady użycia Matplotlib,
- Tworzenie różnych rodzajów wykresów,
- Zaawansowane przykłady z wizualizacją 3D i animacjami.

```python

import matplotlib.pyplot as plt

# Dane
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Tworzenie wykresu
plt.plot(x, y, label='Funkcja y(x)', marker='o')
plt.title('Przykładowy wykres liniowy')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.legend()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Generowanie danych
data = np.random.randn(1000)

# Histogram
plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.title('Histogram danych losowych')
plt.xlabel('Wartości')
plt.ylabel('Częstość')
plt.show()
```

```python

import matplotlib.pyplot as plt

# Dane
x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 8, 7]

# Tworzenie wykresu punktowego
plt.scatter(x, y, color='red', label='Punkty')
plt.title('Przykładowy wykres punktowy')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Dane
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Wykres 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title('Wykres 3D')
plt.show()
```

```python

import matplotlib.pyplot as plt

# Dane
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Tworzenie wielu wykresów
plt.figure(figsize=(10, 5))

# Pierwszy wykres
plt.subplot(1, 2, 1)
plt.plot(x, y1, label='Kwadrat', color='blue')
plt.title('Wykres kwadratu')
plt.legend()

# Drugi wykres
plt.subplot(1, 2, 2)
plt.plot(x, y2, label='Liniowy', color='red')
plt.title('Wykres liniowy')
plt.legend()

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Dane
x = np.linspace(0, 2 * np.pi, 1000)

# Tworzenie figury
fig, ax = plt.subplots()
line, = ax.plot(x, np.sin(x))

# Funkcja aktualizacji
def update(frame):
    line.set_ydata(np.sin(x + frame / 10))
    return line,

# Animacja
ani = FuncAnimation(fig, update, frames=100, interval=50)
#plt.show()
# Wyświetlenie w Jupyter Notebook
HTML(ani.to_jshtml())
```
